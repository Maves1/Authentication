```python
import numpy as np

dataset_path = 'hmog_public_dataset/'
```

```python
import pickle

test_file = open('preprocessed_data/hmog_dataset_ww_ws', 'rb')
session_hmog_matrices = pickle.load(test_file)
test_file.close()
```

```python
# We need to exclude subjects who used landscape orientation
landscape_subjects = ['856302', '751131', '220962', '186676', '980953', '201848', '675397']
for l_s in landscape_subjects:
    del session_hmog_matrices[l_s]

subjects_list = list(session_hmog_matrices.keys())
print(f'len: {len(subjects_list)}\n\n{subjects_list}')
```

```python
# 1. Let's add a label (walking or sitting - w or s) to each hmog authentication vector
# 2. Let's concatenate all sessions into a single array of hmog features
hmog_vectors_with_labels_list = []

for subject in subjects_list:
    for session_type in session_hmog_matrices[subject].keys():
        for i in range(len(session_hmog_matrices[subject][session_type])):
            session = session_hmog_matrices[subject][session_type][i]
            
            curr_label = 1 if session_type == 'w' else 0
            session = np.concatenate((session, np.full( (session.shape[0], 1), curr_label, dtype=np.int32) ), axis=1)
            
            hmog_vectors_with_labels_list.append(session)
            print(hmog_vectors_with_labels_list[-1].shape)
            
hmog_vectors_with_labels = np.concatenate(hmog_vectors_with_labels_list, axis=0)
    
```

Now we have a suitable dataset with labels
